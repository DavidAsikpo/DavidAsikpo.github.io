```python
daily_tickets = data.groupby(data['ticket_created_date']).size().reset_index(name = "ticket_count")
```

Rolling 7 Day Average Analysis 

```python
daily_tickets['rolling_7'] = daily_tickets['ticket_count'].rolling(window =7).mean()
daily_tickets
```

```python
x_start = daily_tickets["ticket_created_date"].min()
x_end = daily_tickets["ticket_created_date"].max()

y_min = daily_tickets["rolling_7"].min()
y_max = daily_tickets["rolling_7"].max()


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))

sns.lineplot(data=daily_tickets, x="ticket_created_date", y="rolling_7")

plt.xticks([x_start, x_end])
plt.yticks([y_min, y_max])

plt.title("Smoothed Daily Trend (7-day average)")
plt.xlabel("Date")
plt.ylabel("Ticket Volume")

plt.show()
```

Day Of Week Analysis 

```python
data['day_of_week'] = data['ticket_created_date'].dt.day_of_week
data
```

```python
def daily_filtered():

    day_map = {
        0: "Monday",
        1: "Tuesday",
        2: "Wednesday",
        3: "Thursday",
        4: "Friday",
        5: "Saturday",
        6: "Sunday"}

    for i in range(0,7):
        filtered_ = data[data['day_of_week'] == i]
        group_ = filtered_.groupby(filtered_['ticket_created_date']).size().reset_index(name = "ticket_count")

        group_["ticket_created_date"] = pd.to_datetime(group_["ticket_created_date"])
        group_["ticket_count"] = group_["ticket_count"].astype(int)

        x_start = group_["ticket_created_date"].min()
        x_end = group_["ticket_created_date"].max()

        y_min = group_["ticket_count"].min()
        y_max = group_["ticket_count"].max()


        import matplotlib.pyplot as plt
        import seaborn as sns

        plt.figure(figsize=(14,6))

        sns.lineplot(data= group_, x="ticket_created_date", y="ticket_count")

        plt.xticks([x_start, x_end])
        plt.yticks([y_min, y_max])

        plt.title(f"{day_map[i]} Smoothed Daily Trend")
        plt.xlabel("Date")
        plt.ylabel("Ticket Volume")

        plt.show()


```

```python
snapshot_1 = data[(data['status'] == "Closed") | (data['status'] == "Resolved")]
snapshot_closed = snapshot_1.groupby(snapshot_1["ticket_created_date"]).size().reset_index(name = "tickets resolved")
snapshot_2 = data[(data['status'] == "Open") | (data['status'] == "In Progress")]
snapshot_open = snapshot_2.groupby(snapshot_2["ticket_created_date"]).size().reset_index(name = "tickets not resolved")
snapshot_3 = data[(data['status'] == "Pending Customer")]
snapshot_customer_p = snapshot_3.groupby(snapshot_3["ticket_created_date"]).size().reset_index(name = "tickets pending customer")




```

```python
combined = pd.merge(
    daily_tickets,
    snapshot_closed,
    on="ticket_created_date",
    how="outer"
)

combined = pd.merge(
    combined,
    snapshot_open,
    on="ticket_created_date",
    how="outer"
)

combined = pd.merge(
    combined,
    snapshot_customer_p,
    on="ticket_created_date",
    how="outer"
)

combined
```

```python
plt.figure(figsize=(12,4))
sns.lineplot(data = combined, x= "ticket_created_date", y = "tickets not resolved")
plt.xlabel("ticket date")
plt.ylabel("tickets not resolved")

x_min = combined['ticket_created_date'].min()
x_max = combined['ticket_creted_date'].max()

y_min = combined['tickets not resolved'].min()
y_max = combined['tickets not resolved'].max()

plt.xticks([x_min, x_max])
plt.yticks([y_min, y_max])

plt.show()

```

* Is demand outpacing support capacity?



```python
combined['year_month'] = combined['ticket_created_date'].dt.to_period("M") 
combined_ym = combined.groupby(combined['year_month'])[['ticket_count','tickets resolved','tickets not resolved','tickets pending customer']].sum().reset_index()
combined_ym["year_month"] = combined_ym['year_month'].astype(str)

column_names = ['ticket_count','tickets resolved','tickets not resolved','tickets pending customer']
for names in column_names:
    combined_ym[names] = combined_ym[names].astype(int)

combined_ym['tickets_total'] = combined_ym['ticket_count'] - combined_ym['tickets pending customer']
combined_ym.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

sns.lineplot(data=combined_ym, x="year_month", y="tickets_total", label="Demand")
sns.lineplot(data=combined_ym, x="year_month", y="tickets resolved", label="Resolved")

plt.fill_between(
    combined_ym["year_month"],
    combined_ym["tickets_total"],
    combined_ym["tickets resolved"],
    where=(combined_ym["tickets_total"] > combined_ym["tickets resolved"]),
    color = "red",
    alpha=0.3,
    label="Backlog Growth")


plt.legend()
plt.title("Demand vs Resolution Capacity")
plt.xticks(rotation=45)
plt.show()
```
